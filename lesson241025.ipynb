```python
#plotscatter for the three targets
import matplotlib.pyplot as plt
Xnew =  X[y == 0, :]
plt.scatter (Xnew[:, 1], Xnew[:,2], c='r', label='class 0')
# first argument is the feature two, the third argument mean take the column 3 as axis y that means take the feature 3
Xnew =  X[y == 1, :]
plt.scatter (Xnew[:, 1], Xnew[:,2], c='b', label='class 1')


Xnew =  X[y == 2, :] #means take the rows where y ==2
plt.scatter (Xnew[:, 1], Xnew[:,2], c='g', label='class 2')

plt.title('iris dataset')
plt.xlabel('feature 2')
plt.ylabel('feature 3')
plt.legend(loc='best')
```
