```python
bins= [0, 5, 11, 100]
level_of_experiences = pd.cut(df_dipendents['Anni di esperienza'].values, bins, right=False, labels = ['junior', 'mid', 'senior'])
level_of_experiences
```

```python
def find_level(anni):
    if anni < 5:
        return 'Junior'
    elif anni <= 10:
        return 'Mid'
    else:
        return 'Senior'

df_dipendents['level_of_experiences'] = df_dipendents['Anni di esperienza'].map(find_level)
```
