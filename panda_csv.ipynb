Removing Duplicates

```python
data2 = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                      'k2': [1, 1, 2, 3, 3, 4, 4]})
data2
```

```python
data2.duplicated() #if the rows is duplicated
```

```python
data2.drop_duplicates()
```

```python
data = pd.DataFrame({'device': ['iphone', 'galaxy', 'ipad'], 'price': [987, 1045, 680]})
data
```

```python
device_brand = {'iphone': 'Apple', 'ipad': 'Apple', 'galaxy': 'Samsung'}
```

```python
#new column with the info from dict

data['brand'] = data['device'].str.lower().map(device_brand)
data
```
