```python
plt.subplot(1,3,1) #1rows and 3 columns in the grid so the graph it will be in 1
plt.plot([1,2,3,4], [1,4,9,16], 'r--')

plt.subplot(1,3,3) #1rows and 3 columns in the grid so the graph it will be in 1
plt.plot([1,2,3,4], [11,14,19,16], 'y--')


plt.subplot(1,3,2) #1rows and 3 columns in the grid so the graph it will be in 1
plt.plot([11,12,13,14], [1,4,9,16], 'g--')
```

Scatter Plots

```python
data1 = 20* np.random.randn(100) + 100
data2 = data1 + 50
plt.scatter(data1, data2, s=100, c='pink')
plt.xlabel('Dim1')
plt.ylabel('Dim2')
```
