```python
import plotly.express as px
```

Bar plot

```python
df = px.data.gapminder().query("continent == 'Europe' and year == 2007 and pop > 2.e6")
fig = px.bar(df, y='pop', x='country', text_auto='.2s',
             title="various text sizes")
fig.show()
```

```python
df2 = px.data.carshare()
fig1 = px.scatter_map(df2, lat="centroid_lat", lon="centroid_lon", color='peak_hour', size="car_hours", map_style="carto-positron", size_max=15, zoom=10, color_continuous_scale=px.colors.cyclical.IceFire)
fig1.show()
```
