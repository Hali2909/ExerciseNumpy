```python
import pandas as pd
import matplotlib as mpl
import numpy as np

fd = pd.read_csv("weather.csv", sep=";")

fd.head(5)
```

```python
#number rows and number columns


print(f'Number column {fd.shape[1]}')
print(f'Number rows {fd.shape[0]}')
```

```python
fd.isnull().sum()
print(f'Sum of null values: {fd.isnull().sum().sum()}')
```

```python
#add new columns

fd["temp_diff"] = abs(fd['Temp9am']-fd["Temp3pm"])
fd["hum_diff"] = abs(fd['Humidity9am']-fd["Humidity3pm"])
```

```python
temphum = fd[(fd["temp_diff"] > 10) & (fd["hum_diff"] > 20)]
grouped = temphum.groupby("Location")
grouped
```

```python
fd[(fd["temp_diff"] > 10) & (fd["hum_diff"] > 20)]["Location"].value_counts().idxmax()
```

```python
#number of locations
fd["Location"].nunique()
```

```python
fd["RainToday"] = fd["RainToday"].map({'Yes': 1, 'No': 0})
rain_counts = fd.groupby("Location")["RainToday"].sum()
print(rain_counts)
```
