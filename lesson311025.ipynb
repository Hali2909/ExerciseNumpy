Metrics

```python
#if the target is not balanced it's better to use recall, precision, f1score instead of the accuract
#ROC

import numpy as np
from sklearn.metrics import roc_auc_score

y_true = np.array([0,0,1,1])
y_score = np.array([0.1, 0.4, 0.35, 0.8])
roc_auc_score(y_true, y_score)

```

Confusion_matrix
in her principal diagonal line there is the value of the positive true for each sample

```python
from sklearn.metrics import confusion_matrix

y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
confusion_matrix(y_true, y_pred)
```
