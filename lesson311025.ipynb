Metrics for the classifier

```python
#if the target is not balanced it's better to use recall, precision, f1score instead of the accuract
#ROC

import numpy as np
from sklearn.metrics import roc_auc_score

y_true = np.array([0,0,1,1])
y_score = np.array([0.1, 0.4, 0.35, 0.8])
roc_auc_score(y_true, y_score)

```

Confusion_matrix
in her principal diagonal line there is the value of the positive true for each sample

```python
from sklearn.metrics import confusion_matrix

y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
confusion_matrix(y_true, y_pred)
```

Classification report

```python
from sklearn.metrics import classification_report

y_true = [0, 1, 2, 2, 0]
y_pred = [0, 0, 2, 1, 0] 

target_names = ['class 0', 'class 1', 'class 2']

print(classification_report(y_true, y_pred, target_names=target_names))
```

Regression metrics

MEAN ABSOLUTE ERROR

```python
from sklearn.metrics import mean_absolute_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 12]

mean_absolute_error(y_true, y_pred) #avg error
```

Root mean square error

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 12]

sqrt(mean_squared_error(y_true, y_pred))
```

```python
#MEAN SQUARED LOGARITHMIC ERROR

from sklearn.metrics import mean_squared_log_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 12]

mean_squared_log_error(y_true, y_pred) 
```

```python
#R^2 score
#how the value change basend on the average
from sklearn.metrics import r2_score
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 12]

r2_score(y_true, y_pred) #avg error
```

Dummy estimators

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X, y = iris.data, iris.target
y[y != 1] = -1
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

clf = SVC(kernel='linear', C=1).fit(X_train, y_train)
print(clf.score(X_test, y_test))
#now we are going to use the dummy to see if svc is better(this is just a confront to see if our algorith is better than the dummy one)
clf = DummyClassifier(strategy='uniform', random_state=0)
clf.fit(X_train, y_train)
result = clf.predict(X_test)
print(accuracy_score(result, y_test))
#given that the difference is not big now we know that we should change the model or his iperparameters
```

```python
clf = SVC(gamma='scale', kernel='rbf', C=1).fit(X_train, y_train)
clf.score(X_test,y_test)
```

HOW CAN WE KNOW WHICH IS THE BEST PARAMETERS 

CROSS VALIDATION GIVE US THE OPPORTUNITY TO EXPERIMENT THE BEST PARAMETERS

The cross validation divide the data in train and test and then it split the train part leaving some random part of data out of the training to use them then as a test.

Better explaination from copilot
- Process:
- The dataset is split into training and testing parts.
- The training set is further divided into k folds (subsets).
- The model is trained on k-1 folds and validated on the remaining fold.
- This process repeats until each fold has been used as a validation set once.
- The performance is averaged across folds → more reliable estimate


```python
#Let's first create a dictionary with the parameters that i want to try

from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

parameters = {
    'kernel': ('linear', 'rbf', 'poly'),
    'C' : [0.1,1,10]
    }


svc = svm.SVC(gamma='scale')

clf = GridSearchCV(svc,parameters, cv=5) #last argument is cross validation that means the training set is splitted in 5 fold
clf.fit(iris.data, iris.target)
```

```python
clf.best_score_
```

RandomSearch
beyond the gridsearch where we set up the values to use we can use also the randomsearch that use a random value and test random combinations 

We can also train the model every time with a different train set and different test set

```python
from sklearn.model_selection import cross_val_score

clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, iris.data, iris.target, cv=5)
#the model will train 5 times and every time with a different set train(one of the 5)

print(scores)
```

```python
scores.mean() #we can undestarnd how much the model is good in this dataset
```

```python
#if we want the recall instead of accuracy
from sklearn.model_selection import cross_val_score

clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, iris.data, iris.target, cv=5, scoring='recall_macro')

print(scores)
```
